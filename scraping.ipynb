```python
import instaloader
import pandas as pd
from google.colab import files

# Crea un'istanza di Instaloader
L = instaloader.Instaloader()

# Funzione per ottenere i primi 2 post di un profilo senza autenticazione
def get_first_two_posts(profile_name):
    # Carica il profilo
    profile = instaloader.Profile.from_username(L.context, profile_name)

    # Ottieni i primi due post
    posts = profile.get_posts()

    # Lista per memorizzare i dati dei post
    post_data = []

    for i, post in enumerate(posts):
        if i >= 3:
            break
        post_info = {
            'profile_name': profile_name,
            'data_pubblicazione': post.date,
            'descrizione': post.caption,
            'numero_like': post.likes,
            'numero_commenti': post.comments,
            'url_immagine': post.url
        }
        post_data.append(post_info)

    return post_data

# Lista dei profili Instagram
profiles = [
    'artpanorama_',
    'mistral_beach_club',
    'everyday_pmo',
    'remember.pmo',
    'magazine.events',
    'noname_sicily',
    '_urbanparty',
    'apericool',
    'essentials.events',
    'elle.eventi',
    'theweekendpalermo',
    'tnos_eventi',
    'unlockedparty',
    'timelab_cefalu',
    'fireworks_project_',
    'citysea_rooftop',
    'popshock_',
    'habitat__organization',
    'revup_official',
    'disordine_palermo',
    'savage.events_',
    'hangover_pa',
    'houselab_events',
    'cyberlab.comm',
    'eventi_paradise',
    'traphouse.th',
    'highday.pmo',
    'risolutore.adrenalina',
    'bigreunion_',
    'mindhousepalermo',
    'eden_club',
    'icandelai',
    'officina_palermo',
    'atmosferaevent',
    'novaserviceista',
    '00_open',
    '90100_studio',
    'muevelo.sicily',
    'wave_pmo',
    'albertinibeach',
    'nocap.__events',
    'nked.pmo',
    'panorama.eventi'
]

# Lista per raccogliere tutti i dati dei post
all_post_data = []

# Esegui scraping per ogni profilo
for profile_name in profiles:
    try:
        post_data = get_first_two_posts(profile_name)
        all_post_data.extend(post_data)
    except Exception as e:
        print(f"Errore durante lo scraping di {profile_name}: {e}")

# Crea un DataFrame con tutti i dati dei post
df = pd.DataFrame(all_post_data)

# Salva i dati in un file Excel
filename = 'instagram_posts.xlsx'
df.to_excel(filename, index=False)

# Scarica il file Excel
files.download(filename)
```
